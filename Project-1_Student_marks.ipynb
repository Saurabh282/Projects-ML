```python
import pandas as pd
```

```python
df_student_marks=pd.read_csv('student_marks.csv')
df_student_marks
```

```python
df_student_marks.dtypes
```

```python
#Create a new column of total marks
total_marks=pd.Series()
total_marks=0
subjects=['Maths','Physics','Chemistry','English','Biology','Economics','History','Civics']
for subject in subjects:
    total_marks=total_marks+df_student_marks[subject]
```

```python
df_student_marks['total_marks']=total_marks
```

```python
#Check for distinction in maths
df_student_marks.loc[df_student_marks['Maths']>=75,['Name','Gender','Maths']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# total marks of male student vs total marks of female student
sns.barplot(x=df_student_marks['Gender'],y=df_student_marks['total_marks'],data=df_student_marks)
plt.show()
```

```python
#Comparing total marks of each student
sns.barplot(x=df_student_marks['Name'],y=df_student_marks['total_marks'],data=df_student_marks)
plt.rcParams['figure.figsize']=(8,6)
plt.show()
```
