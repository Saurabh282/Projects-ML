```python
#Now separating input and output
df_x=df1.drop(columns=['num'])
y=df1[['num']]
```

```python
df_x.head()
```

```python
y.head()
```

```python
# Scaling features
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df_x)
x=sc.transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        #print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
#lets make a function which evaluates the model using cross_val_score
from sklearn.model_selection import cross_val_score
def model_evaluation(model,x,y):
    c_scores=cross_val_score(model,x,y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print(c_scores)
```

```python
#Lets use decision tree
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
r_state=maxr2_score(dtr,x,y)
```

```python
#lets use random forest regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,100,500]}
clf = GridSearchCV(rfr, parameters, cv=5,scoring="r2")
clf.fit(x,y)
clf.best_params_
```

```python
rfr=RandomForestRegressor(n_estimators=500)
r_state=maxr2_score(rfr,x,y)
```

```python
#lets check KNN 
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
parameters = {"n_neighbors":range(2,30)}
clf = GridSearchCV(knr, parameters, cv=5,scoring="r2")
clf.fit(x,y)
clf.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=13)
r_state=maxr2_score(knr,x,y)
```

```python
#Lets use SVM regressor
from sklearn.svm import SVR
import warnings
warnings.filterwarnings("ignore")
svr=SVR()
parameters={"kernel":["linear", "poly", "rbf"],"C":[0.001,0.01,0.1,1,10]}
clf = GridSearchCV(svr, parameters, cv=5,scoring="r2")
clf.fit(x,y)
clf.best_params_
```

```python
svr=SVR(kernel="linear",C=0.1)
r_state=maxr2_score(svr,x,y)
```

## Now lets do cross validation

```python
print("DECISION TREE REGRESSOR\n\n")
model_evaluation(dtr,x,y)
```

```python
print("RANDOM FOREST REGRESSOR\n\n")
model_evaluation(rfr,x,y)
```

```python
print("KNN REGRESSOR\n\n")
model_evaluation(knr,x,y)
```

```python
print("SVM REGRESSOR\n\n")
model_evaluation(svr,x,y)
```

```python
#lets make model using SVM
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 71,test_size=0.20)
svr.fit(x_train,y_train)
y_pred = svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("RMSE is: ",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score is: ",r2_score(y_test,y_pred))
```
