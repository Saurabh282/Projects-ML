```python
# we have categorical feature: region and type
# region has 54 type in it so it difficult to transform it in label encode so i will drop it
#and in "type" here is only two type of category so it is easy to transform in label encode

# Now we have year, month and day columns in dataset so i decided to drop 'Date' column also.
```

```python
df_final=pd.get_dummies(df_av.drop(['region','Date'],axis=1),drop_first=True)
```

```python
df_final.head()
```

```python
#now separate input and output 
df_x=df_final.iloc[:,1:-1]
y=df_final['AveragePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
import numpy as np
from sklearn import metrics
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
pred=dtr.predict(x_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(x_train,y_train)
pred=rfr.predict(x_test)
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

Here we can see that RMSE of Random Forest Regressor is lower than Decision Tree Regressor and Linear regression

```python
#Lets save the model
from sklearn.externals import joblib
joblib.dump(rfr,'Average_price_avocado.pkl')
```
