```python
# treating the skewness
import numpy as np
for col in df_baseball.columns:
    if df_baseball.skew().loc[col]>0.5:
        df_baseball[col]=np.sqrt(df_baseball[col])
    if df_baseball.skew().loc[col]<-0.5:
        df_baseball[col]=np.cbrt(df_baseball[col])
```

```python
# Now separating input and target data
df_x=df_baseball.drop(columns=["W"])
y=df_baseball[["W"]]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test = train_test_split(df_x,y,random_state= r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ", r_state,"is",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state,"is",max_r_score)
    return final_r_state
```

```python
# Cross validation
from sklearn.model_selection import cross_val_score
def model_evaluation(model,df_x,y):
    c_scores=cross_val_score(model,df_x,y,cv=5,scoring='r2')
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("Standard deviation in r2 score for regressor: ",c_scores.std())
```

```python
#Decision tree 
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
r_state=maxr2_score(dtr,df_x,y)
```

```python
# Linear regression
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
r_state=maxr2_score(lreg,df_x,y)
```

```python
#Cross validation of models
```

```python
print("Decision Tree Regressor\n\n")
model_evaluation(dtr,df_x,y)
```

```python
print("Linear Regression\n\n")
model_evaluation(lreg,df_x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_x,y,random_state=90,test_size=0.20)
lreg=LinearRegression()
lreg.fit(x_train,y_train)
y_pred=lreg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print('r2 score is: ',r2_score(y_test,y_pred))
print('RMSE is: ',np.sqrt(mean_squared_error(y_test,y_pred)))
```
