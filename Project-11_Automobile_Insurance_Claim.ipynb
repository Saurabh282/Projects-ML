```python
#use labelencoder to convert all the object type columns to int
from sklearn.preprocessing import LabelEncoder
for col in df.columns:
    if df[col].dtype=="object":
        df[col]=LabelEncoder().fit_transform(df[col])
```

```python
# now check the dataset
df.head()
```

```python
# Now separating dataset into input and output
df_x=df.drop(columns=["Claim Amount"])
y=df[["Claim Amount"]]
```

```python
# To find random state which gives maximum r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        #print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
#lets make a function which evaluates the model using cross_val_score
from sklearn.model_selection import cross_val_score
def model_evaluation(model,df_x,y):
    c_scores=cross_val_score(model,df_x,y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print(c_scores)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
r_state=maxr2_score(dtr,df_x,y)
```

```python
# Cross validation score
print("Decision Tree Regressor\n\n")
model_evaluation(dtr,df_x,y)
```

```python
# random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,100,500]}
clf = GridSearchCV(rfr, parameters, cv=5,scoring="r2")
clf.fit(df_x,y)
clf.best_params_
```

```python
rfr=RandomForestRegressor(n_estimators=500)
r_state=maxr2_score(rfr,df_x,y)
```

```python
# Cross validation score
print("Random Forest Regressor\n\n")
model_evaluation(rfr,df_x,y)
```

```python
# KNN regressor
from sklearn.neighbors import KNeighborsRegressor
knr=KNeighborsRegressor()
parameters = {"n_neighbors":range(2,30)}
clf = GridSearchCV(knr, parameters, cv=5,scoring="r2")
clf.fit(df_x, y)
clf.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=22)
r_state=maxr2_score(knr,df_x,y)
```

```python
# Cross validation score
print("KNN Regressor\n\n")
model_evaluation(knr,df_x,y)
```

```python
#lets make model using Random Forest Regressor
x_train, x_test, y_train, y_test = train_test_split(df_x,y,random_state = 97,test_size=0.20)
rfr.fit(x_train,y_train)
y_pred = rfr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("RMSE is: ",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r2_score is: ",r2_score(y_test,y_pred))
```
