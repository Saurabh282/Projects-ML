```python
# Lets scale the column features into StandardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(df_X)
x=sc.transform(df_X)
x=pd.DataFrame(x,columns=df_X.columns)
```

```python
# use pca to reduce the dimensionality
from sklearn.decomposition import PCA
pca = PCA(n_components=25)
x_pca=pca.fit_transform(x)
```

```python
x_pca.shape
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
```

```python
from sklearn.metrics import f1_score
```

```python
def maxf1_score(clf,x,y):
    maxf=0
    rs=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state = r_state,test_size=0.30,stratify=y)
        clf.fit(x_train,y_train)
        y_pred=clf.predict(x_test)
        f1score=f1_score(y_test,y_pred)
        print("random state :",r_state," and f1 score: ",f1score)
        if f1score>maxf:
            maxf=f1score
            rs=r_state
    print("maximum f1_score is at random state :",rs," and it is :",maxf)
```

```python
# Logistic Regreession
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
lg=LogisticRegression()
maxf1_score(lg,x_pca,y)
```

```python

print(cross_val_score(lg,x_pca,y,cv=5,scoring="f1"))
print("Mean of f1 score ",cross_val_score(lg,x_pca,y,cv=5,scoring="f1").mean())
print("SD in f1 score ",cross_val_score(lg,x_pca,y,cv=5,scoring="f1").std())
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
maxf1_score(dc,x_pca,y)
```

```python
#
print("Mean of f1 score decision tree ",cross_val_score(dc,x_pca,y,cv=5,scoring="f1").mean())
print("sd of f1 score decision tree ",cross_val_score(dc,x_pca,y,cv=5,scoring="f1").std())
print(cross_val_score(dc,x_pca,y,cv=5,scoring="f1"))
```

```python
# Now use random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
parameters={"n_estimators":[10,100,500]}
rf_clf=RandomForestClassifier()
clf = GridSearchCV(rf_clf, parameters, cv=5,scoring="f1")
clf.fit(x_pca,y)
clf.best_params_
```

```python
rf_clf=RandomForestClassifier(n_estimators=500)
maxf1_score(rf_clf,x_pca,y)
```

```python
#
print(cross_val_score(rf_clf,x_pca,y,cv=5,scoring="f1"))
print("Mean of f1 score ",cross_val_score(rf_clf,x_pca,y,cv=5,scoring="f1").mean())
print("SD in f1 score ",cross_val_score(rf_clf,x_pca,y,cv=5,scoring="f1").std())
```

```python
from sklearn.ensemble import GradientBoostingClassifier
parameters={"learning_rate":[0.001,0.01,0.1,1],"n_estimators":[10,100,500,1000]}
gb_clf=GradientBoostingClassifier()
clf = GridSearchCV(gb_clf, parameters, cv=5,scoring="f1")
clf.fit(x_pca,y)
clf.best_params_
```

```python
gb_clf=GradientBoostingClassifier(learning_rate=0.1,n_estimators=500)
maxf1_score(gb_clf,x_pca,y)
```
