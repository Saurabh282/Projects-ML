```python
x=df_mushroom.iloc[:,1:23]
x.shape
```

```python
y=df_mushroom.iloc[:,0]
y.shape
```

```python
y
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
x1=sc.transform(x)
x=pd.DataFrame(x1,columns=x.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(y)
y=le.transform(y)
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=55)
p_c = pca.fit_transform(x) #p_c =Principal_components
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import f1_score
```

```python
def maxf1_score(clf,x,y):
    maxf=0
    rs=0
    for r_state in range(42,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state = r_state,test_size=0.20,stratify=y)
        clf.fit(x_train,y_train)
        y_pred=clf.predict(x_test)
        f1score=f1_score(y_test,y_pred)
        print("random state :",r_state," and f1 score: ",f1score)
        if f1score>maxf:
            maxf=f1score
            rs=r_state
    print("maximum f1_score is at random state :",rs," and it is :",maxf)
```

```python
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")
lg=LogisticRegression()
maxf1_score(lg,p_c,y)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(lg,p_c,y,cv=5,scoring="f1"))
print("Mean of f1 score ",cross_val_score(lg,p_c,y,cv=5,scoring="f1").mean())
print("SD in f1 score ",cross_val_score(lg,p_c,y,cv=5,scoring="f1").std())
```

```python
from sklearn.tree import DecisionTreeClassifier
dc=DecisionTreeClassifier()
maxf1_score(dc,p_c,y)
```

```python
from sklearn.model_selection import cross_val_score
print("Mean of f1 score decision tree ",cross_val_score(dc,p_c,y,cv=5,scoring="f1").mean())
print("sd of f1 score decision tree ",cross_val_score(dc,p_c,y,cv=5,scoring="f1").std())
print(cross_val_score(dc,p_c,y,cv=5,scoring="f1"))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
svc=SVC()
parameters={"kernel":["linear", "poly", "rbf"],"C":[0.001,0.01,0.1,1,10]}
clf = GridSearchCV(svc, parameters, cv=5,scoring="f1")
clf.fit(p_c,y)
clf.best_params_
```

```python
svc=SVC(kernel="poly",C=0.1)
maxf1_score(svc,p_c,y)
```

```python
print("Mean of f1 score for SVM classifier: ",cross_val_score(svc,p_c,y,cv=5,scoring="f1").mean())
print("standard deviation in f1 score for SVM classifier: ",cross_val_score(svc,p_c,y,cv=5,scoring="f1").std())
print(cross_val_score(svc,p_c,y,cv=5,scoring="f1"))
```

```python
svc=SVC(kernel="poly",C=0.1)
x_train,x_test,y_train,y_test=train_test_split(p_c, y,random_state = 45,test_size=0.20,stratify=y)
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
print("Confusion matrix \n ",confusion_matrix(y_test,y_pred))
print("f1 score is ",f1_score(y_test,y_pred))
print("classification report \n",classification_report(y_test,y_pred))
print("AUC ROC Score: ",roc_auc_score(y_test,y_pred))
```
